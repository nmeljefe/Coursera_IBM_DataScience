Use the Geospatial Coordinates csv file to generate Latitude and Longitude coordinates by Postal Code:

```python
import csv

with open ('Geospatial_Coordinates.csv', newline = '') as f:
    reader = csv.reader(f)
    PC_Coords = list(reader)[1:]

print(PC_Coords)
print(len(PC_Coords))
```

```python
df_LatLng = pd.DataFrame(PC_Coords, columns =['Postal Code', 'Latitude', 'Longitude'], dtype = float) 
df_LatLng.head(105)
```

Now use pandas to merge the two dataframes together using the Postal Code value and an inner join:

```python
dfinal = df.merge(df_LatLng, on="Postal Code", how = 'inner')
dfinal.head(105)
```

Now for the visualizations.


```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(dfinal['Borough'].unique()),
        dfinal.shape[0]
    )
)
```

Use geopy and geolocater to get coordinates for Toronto, ON.  Call the user_agent "toronto_explorer":

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto, Ontario are {}, {}.'.format(latitude, longitude))
```

Now create a map of Toronto neighborhoods superimposed:

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(dfinal['Latitude'], dfinal['Longitude'], dfinal['Borough'], dfinal['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
